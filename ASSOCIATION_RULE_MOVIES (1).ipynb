```python
data=pd.read_csv('my_movies.csv')
data
```

```python
data.columns
```

```python
movies = data.iloc[:,5:15]
movies
```

#min support = 0.1
#hyperparameter = confidence

```python
items=apriori(movies,min_support=0.1,use_colnames=True)
items
```

```python
rules= association_rules(items,metric='confidence',min_threshold=0.9)
rules
```

```python
rules.sort_values('confidence',ascending=False)
```

```python
plt.scatter(data=rules,x='support',y='confidence')
```

#min support =0.1
#hyperparameter =lift

```python
rules_lift = association_rules(items,metric='lift',min_threshold=0.9)
rules_lift
```

```python
rules_lift.sort_values('lift',ascending=False)
```

```python
plt.scatter(data=rules_lift,x='support',y='confidence')
```
